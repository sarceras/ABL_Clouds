# Data


```python
# Define function that loads a .mat file converting it into a pandas.DataFrame.
def load_mat_to_dataframe(path_to_mat, index = "latx", columns = ["q0", "thv0", "yq", "ythv"]):
    # Load the data.
    data = loadmat(path_to_mat)
    # Create DataFrame.
    df = pd.DataFrame(index = data[index].flatten())
    # Set the name to the DataFrame index.
    df.index.name = "Latitude"
    # Add the columns to the DataFrame.
    for column in columns:
        df[column] = data[column]
    return df
```

```python
# Load the Grass data for each season.
datas_grass = {1: load_mat_to_dataframe("Free Atmosphere/Wet/Grass/DJF.mat"), # Winter (DJF).
               2: load_mat_to_dataframe("Free Atmosphere/Wet/Grass/MAM.mat"), # Spring (MAM).
               3: load_mat_to_dataframe("Free Atmosphere/Wet/Grass/JJA.mat"), # Summer (JJA).
               4: load_mat_to_dataframe("Free Atmosphere/Wet/Grass/SON.mat")} # Autumn (SON).
```

```python
# Load the Forest data for each season.
datas_forest = {1: load_mat_to_dataframe("Free Atmosphere/Wet/Forest/DJF.mat"), # Winter (DJF).
                2: load_mat_to_dataframe("Free Atmosphere/Wet/Forest/MAM.mat"), # Spring (MAM).
                3: load_mat_to_dataframe("Free Atmosphere/Wet/Forest/JJA.mat"), # Summer (JJA).
                4: load_mat_to_dataframe("Free Atmosphere/Wet/Forest/SON.mat")} # Autumn (SON).
```

```python
# Load the ForestGrass data for each season.
datas_forestgrass = {1: load_mat_to_dataframe("Free Atmosphere/Wet/ForestGrass/DJF.mat"), # Winter (DJF).
                     2: load_mat_to_dataframe("Free Atmosphere/Wet/ForestGrass/MAM.mat"), # Spring (MAM).
                     3: load_mat_to_dataframe("Free Atmosphere/Wet/ForestGrass/JJA.mat"), # Summer (JJA).
                     4: load_mat_to_dataframe("Free Atmosphere/Wet/ForestGrass/SON.mat")} # Autumn (SON).
```

```python
# Map dictionary for seasons.
dict_seasons = {1: "DJF", 2: "MAM", 3: "JJA", 4: "SON"}
# Map dictionary for label fa params.
dict_fa_name = {1: "$q_0$", 2: "$\u03B8_{v0}$", 3: "$\gamma_{q}$", 4: "$\gamma_{\u03B8}$"}
```

```python
# Concatenate Grass data.
df_grass = pd.DataFrame()
for season in datas_grass.keys():
    # Keep the dataframe corresponding to the current season.
    df = datas_grass[season].copy()
    # Add the information of the corresponding type and season to the column names.
    df.columns = pd.MultiIndex.from_product([["Grass"], [season], df.columns], names = ["Specie", "Season", "FA param"])
    # Concatenate data of the various seasons.
    df_grass = pd.concat([df_grass, df], axis = 1).sort_index(axis = 1)

df_grass.head()
```

```python
# Concatenate Forest data.
df_forest = pd.DataFrame()
for season in datas_forest.keys():
    # Keep the dataframe corresponding to the current season.
    df = datas_forest[season].copy()
    # Add the information of the corresponding type and season to the column names.
    df.columns = pd.MultiIndex.from_product([["Forest"], [season], df.columns], names = ["Specie", "Season", "FA param"])
    # Concatenate data of the various seasons.
    df_forest = pd.concat([df_forest, df], axis = 1).sort_index(axis = 1)
    
df_forest.head()
```

```python
# Concatenate ForestGrass data.
df_forestgrass = pd.DataFrame()
for season in datas_forestgrass.keys():
    # Keep the dataframe corresponding to the current season.
    df = datas_forestgrass[season].copy()
    # Add the information of the corresponding type and season to the column names.
    df.columns = pd.MultiIndex.from_product([["ForestGrass"], [season], df.columns], names = ["Specie", "Season", "FA param"])
    # Concatenate data of the various seasons.
    df_forestgrass = pd.concat([df_forestgrass, df], axis = 1).sort_index(axis = 1)
    
df_forestgrass.head()
```

```python
# Concatenate data. #removed forestgrass from here, if you want to add put , df_forestgrass
df = pd.concat([df_grass, df_forest], axis = 1).sort_index(axis = 1)
df.head()
```

## Smooth data

```python
# Parameters for data smooth.
window_lenght = 15 # Needs to be odd.
kpol = 3 # Less than window_lenght.
```

```python
# Keep the dataframe until the 65-th latitude in order to not have nan values.
df_no_nan = df.loc[:65]
# Add a new level column identifier.
df_no_nan.columns = pd.MultiIndex.from_tuples(list(map(lambda x: tuple(list(x) + ["No smooth"]), df_no_nan.columns)), 
                                              names = ["Specie", "Season", "FA param", "Type"])
df_no_nan.head()
```

```python
# Smooth each column of the dataframe.
df_no_nan_smooth = df_no_nan.apply(lambda x: savgol_filter(x, window_lenght, kpol))
# Add a new level column identifier.
df_no_nan_smooth.rename({"No smooth": "Smooth"}, axis = 1, level = "Type", inplace = True)
df_no_nan_smooth.head()
```

```python
# Concatenate two information into one dataframe.
df = pd.concat([df_no_nan, df_no_nan_smooth], axis = 1).sort_index(axis = 1)
df.head()
```

```python
# Example.
label = ("Forest", 2, "q0")

fig, ax = plt.subplots(figsize = (5, 3))
# Plot.
df[label].plot(ax = ax)
# Set attributes of the plot.
ax.set_ylabel(label)
ax.legend();
```

```python
# Keep the unique seasons and fa parameters.
SEASONS = df.columns.get_level_values("Season").unique()
FA_PARAMS = df.columns.get_level_values("FA param").unique()
```

```python
# Smooth data and plot forest-grass-forestgrass, with smooth, and for each season
fig, ax = plt.subplots(4, 4, sharex = True, figsize = (10, 10))
fig.subplots_adjust(wspace = .5, hspace = .4)

subdf = df.xs("Smooth", axis = 1, level = "Type")

for i,season in enumerate(SEASONS):
    # Keep the subdataframe with the current season.
    subdf1 = subdf.xs(season, level = "Season", axis = 1, drop_level = True)
    for j,fa_param in enumerate(FA_PARAMS):
        # Keep the subdataframe with the current fa params.
        subdf2 = subdf1.xs(fa_param, level = "FA param", axis = 1, drop_level = True)
        # Plot lines.
        subdf2.plot(ax = ax[i,j], legend = False)     
        # Set attributes of the plot.
        ax[i,j].set_ylabel(dict_fa_name[j+1])
        ax[i,j].set_title(dict_seasons[season])
        ax[i,j].ticklabel_format(style = "sci", axis = "y", scilimits = (0, 0))

# Add unique legend.
ax.flatten()[-2].legend(bbox_to_anchor = (1.3, -0.5), ncol = 3)

fig.savefig(PATH_SAVE_FIGURES + "figure2.png", bbox_inches = "tight", dpi = 1200)
```

```python
# Smooth data and plot forest-grass-forestgrass, with smooth, and for each season
fig, ax = plt.subplots(4, 4, sharex = True, figsize = (10, 10))
fig.subplots_adjust(wspace = .5, hspace = .4)

subdf = df.xs("Smooth", axis = 1, level = "Type")

for i,season in enumerate(SEASONS):
    # Keep the subdataframe with the current season.
    subdf1 = subdf.xs(season, level = "Season", axis = 1, drop_level = True)
    for j,fa_param in enumerate(FA_PARAMS):
        # Keep the subdataframe with the current fa params.
        subdf2 = subdf1.xs(fa_param, level = "FA param", axis = 1, drop_level = True)
        subdf2["difference"] = subdf2["Forest"] - subdf2["Grass"]
        # Plot lines.
        subdf2[["difference"]].plot(ax = ax[i,j], legend = False)     
        # Set attributes of the plot.
        ax[i,j].set_ylabel(dict_fa_name[j+1])
        ax[i,j].set_title(dict_seasons[season])
        ax[i,j].ticklabel_format(style = "sci", axis = "y", scilimits = (0, 0))

# Add unique legend.
ax.flatten()[-2].legend(bbox_to_anchor = (0.3, -0.5), ncol = 3)

fig.savefig(PATH_SAVE_FIGURES + "figure3.png", bbox_inches = "tight", dpi = 300)
```
