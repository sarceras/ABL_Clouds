## Load data

```python
# Load data of the free atmosphere.
xls = pd.ExcelFile("./Data/FreeAtmoshere.xlsx")

forest_free_atm = pd.read_excel(xls, "Forest", index_col = 0, header = [0, 1, 2])
grass_free_atm = pd.read_excel(xls, "Grass", index_col = 0, header = [0, 1, 2])
forestgrass_free_atm = pd.read_excel(xls, "ForestGrass", index_col = 0, header = [0, 1, 2])

# Concatenate free atmosphere: 
free_atm = pd.concat({"Grass": grass_free_atm, "Forest": forest_free_atm}, axis = 1, names = ["Specie"])
# Keep smooth data for free atmosphere.
free_atm = free_atm.xs("Smooth", axis = 1, level = "Type")
```

```python
# Load data of the Radiation and Albedo.
xls = pd.ExcelFile("./Data/RadiationAlbedo.xlsx")

radiation = pd.read_excel(xls, "Radiation", index_col = 0, header = [0, 1])
grass_albedo = pd.read_excel(xls, "Albedo_grass", index_col = 0, header = [0, 1])
forest_albedo = pd.read_excel(xls, "Albedo_forest", index_col = 0, header = [0, 1])

# Concatenate albedo.
albedo = pd.concat({"Grass": grass_albedo, "Forest": forest_albedo}, axis = 1, names = ["Specie"])
```

```python
# Map dictionary for seasons.
dict_seasons = {1: "DJF", 2: "MAM", 3: "JJA", 4: "SON"}
```

## Analysis

```python
# Define parameters for scenario.
s = 0.45
cloud_albedo = 0.5

soil = "Loam"
```

```python
# Get the maximum radiation from sunrise to sunset per each latitude and season.
radiation_day = radiation.loc["06:00:00":"18:00:00"]
radiation_day_max = radiation_day.groupby(axis = 1, level = "Season").apply(lambda x: x.max().droplevel("Season"))
radiation_day_max.head()
```

```python
# Get the albedo from sunrise to sunset.
albedo_day = albedo.loc["06:00:00":"18:00:00"]
# Reorder axis.
albedo_day = albedo_day.reorder_levels([0, 2, 1], axis = 1).sort_index(axis = 1, level = [0, 1, 2])
albedo_day = albedo_day.stack(level = "Latitude")
albedo_day = albedo_day.reorder_levels([1, 0], axis = 0).sort_index(axis = 0, level = [0, 1])
albedo_day.head()
```

```python
idx = pd.to_datetime(albedo_day.index.get_level_values("Hour").unique())



tt = np.array(idx.hour + (idx.minute/60))
dt = tt[1] - tt[0]
tt
```

```python
# Get the FA parameters of the species.
free_atm_specie = free_atm.stack("FA param")
free_atm_specie.head()
```

```python
def iterate_specie_season(group):
    # Define current specie and season.
    specie, season = group.name
    # Get the maximum radiation for the current season.
    Q_max = radiation_day_max[season]
    # Get the albedo list for the current specie and season.
    alfa = albedo_day[(specie, season)]

    def iterate_latitude(x, Q_max, alfa, season, specie):
        # Define current latitude.
        lat = x.name
        # Get FA data for the current latitude.
        FA_params = x[(specie, season)].loc[lat].to_dict()
        # Get maximum radiation for the current latitude.
        Q_max_lat = Q_max.loc[lat]
        # Get albedo for the current latitude.
        alfa_lat = alfa.loc[lat].values
        # Compute daily sim for the current latitude.
        out = daily_sim(FA_params, Q_max_lat, s, cloud_albedo, specie, soil, alfa_lat, tt)
        return out

    # Compute daily sim for each latitude for the current specie and season.
    print(f"{specie}, season {season}:")
    tqdm.pandas()
    out = group.groupby(axis = 0, level = "Latitude").progress_apply(lambda x: iterate_latitude(x, Q_max, alfa, season, specie))
    return out
```

```python
tqdm.pandas()
daily_output = free_atm_specie.groupby(axis = 1, level = ["Specie", "Season"]).apply(iterate_specie_season)
# Reorder axis.
daily_output = daily_output.stack("Season").reorder_levels([1, 0], axis = 0).sort_index(axis = 0, level = [0, 1])
```

```python
def create_timeline(t1, t2):
    return np.linspace(t1, t2, num = int((t2-t1)/(3600*dt)))
```

```python
def compute_forcing(x):
    interp_grass = interp1d(x[("Grass", "T")], x[("Grass", "Q")], fill_value = "extrapolate")
    interp_forest = interp1d(x[("Forest", "T")], x[("Forest", "Q")], fill_value = "extrapolate")

    # Find two intersections on x-axis (sunrise and sunset).
    T0, T1 = fsolve(interp_grass, [20000, 70000]) # It is equivalent using Forest.  
    t = create_timeline(T0, T1)

    # RADIATION FORCING.  
    # No cloud.
    area_Q_forest = np.trapz(interp_forest(t), t)
    area_Q_grass = np.trapz(interp_grass(t), t)

    # Cloud.
    # Grass.
    if not np.isnan(x[("Grass", "Crossing_time")]):
        t1 = create_timeline(T0, x[("Grass", "Crossing_time")])
        t2 = create_timeline(x[("Grass", "Crossing_time")], T1)

        area_Q_cloud_grass = np.trapz(interp_grass(t1), t1) + np.trapz((1.-cloud_albedo)*interp_grass(t2), t2)
    else:
        area_Q_cloud_grass = area_Q_grass
    
    # Cloud.
    # Forest.
    if not np.isnan(x[("Forest", "Crossing_time")]):
        t1 = create_timeline(T0, x[("Forest", "Crossing_time")])
        t2 = create_timeline(x[("Forest", "Crossing_time")], T1)

        area_Q_cloud_forest = np.trapz(interp_forest(t1), t1) + np.trapz((1.-cloud_albedo)*interp_forest(t2), t2)
    else:
        area_Q_cloud_forest = area_Q_forest

    area_albedo_cloud_check = area_Q_cloud_forest - area_Q_cloud_grass

    # ALBEDO EFFECT.
    # No cloud.
    area_albedo = np.trapz(interp_forest(t) - interp_grass(t), t) 
    
    # Cloud.
    if not np.isnan(x[("Grass", "Crossing_time")]) and not np.isnan(x[("Forest", "Crossing_time")]):
        DELTA_crossing = (-x[("Forest", "Crossing_time")]/3600. + x[("Grass", "Crossing_time")]/3600.)

        if DELTA_crossing <= 0:
            # Crossing for forest after crossing of grass.
            t1 = create_timeline(T0, x[("Grass", "Crossing_time")])
            t2 = create_timeline(x[("Grass", "Crossing_time")], x[("Forest", "Crossing_time")])
            t3 = create_timeline(x[("Forest", "Crossing_time")], T1)            
            
            area_albedo_cloud = np.trapz(interp_forest(t1) - interp_grass(t1), t1) + np.trapz(interp_forest(t2)-(1.-cloud_albedo)*interp_grass(t2), t2) + np.trapz((1.-cloud_albedo)*interp_forest(t3)-(1.-cloud_albedo)*interp_grass(t3), t3)
        else:
            t1 = create_timeline(T0, x[("Forest", "Crossing_time")])   
            t2 = create_timeline(x[("Forest", "Crossing_time")], x[("Grass", "Crossing_time")])
            t3 = create_timeline(x[("Grass", "Crossing_time")], T1)
            
            area_albedo_cloud = np.trapz(interp_forest(t1)-interp_grass(t1), t1) - np.trapz(interp_grass(t2)-(1.-cloud_albedo)*interp_forest(t2), t2) + np.trapz((1.-cloud_albedo)*interp_forest(t3)-(1.-cloud_albedo)*interp_grass(t3), t3)      
            
    elif np.isnan(x[("Grass", "Crossing_time")]) and not np.isnan(x[("Forest", "Crossing_time")]):
        DELTA_crossing = np.nan
        
        t1 = create_timeline(T0, x[("Forest", "Crossing_time")]) 
        t2 = create_timeline(x[("Forest", "Crossing_time")], T1) 

        area_albedo_cloud = np.trapz(interp_forest(t1)-interp_grass(t1), t1) - np.trapz(interp_grass(t2)-(1.-cloud_albedo)*interp_forest(t2), t2)

    elif not np.isnan(x[("Grass", "Crossing_time")]) and np.isnan(x[("Forest", "Crossing_time")]):
        DELTA_crossing = np.nan
        
        t1 = create_timeline(T0, x[("Grass", "Crossing_time")])
        t2 = create_timeline(x[("Grass", "Crossing_time")], T1)

        area_albedo_cloud = np.trapz(interp_forest(t1)-interp_grass(t1), t1) + np.trapz(interp_forest(t2)-(1.-cloud_albedo)*interp_grass(t2), t2)
    
    else:
        DELTA_crossing = np.nan
        area_albedo_cloud = area_albedo

    # Computation of carbon equivalent from daily mean albedo RF (in W/m^2).
    Carbon_eq = RF_to_Ceq(area_albedo/86400.) 
    Carbon_eq_cloud = RF_to_Ceq(area_albedo_cloud/86400.)

    return pd.Series([DELTA_crossing, area_albedo_cloud, area_albedo_cloud_check, area_albedo, Carbon_eq_cloud, Carbon_eq_cloud], 
                     index = ["DELTA_crossing", "area_albedo_cloud", "area_albedo_cloud_check", "area_albedo", "Carbon_eq", "Carbon_eq_cloud"])

forcing_output = daily_output.apply(compute_forcing, axis = 1)
forcing_output
```
